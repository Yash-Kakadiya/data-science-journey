## `Time Complexity`

````python
number = int(input("enter the number"))

digits = "0123456789"
result = ""
while number != 0:
    result = digits[number % 10] + result
    number = number // 10

print(result)

# Time Complexity: O(n), Where n is the number of digits in the input number.
````

````python
L = [1, 2, 3, 4]

sum = 0
for i in L:
    sum = sum + i

product = 1
for i in L:
    product = product * i

print(sum, product)

# Time Complexity: O(n), Where n is the number of elements in list L.
````

````python
A = [1, 2, 3, 4]
B = [5, 6, 7, 8]
for i in A:
    for j in B:
        print(i, j)

# Time Complexity: O(n × m), Where n = length of list A, m = length of list B
# If both lists are of equal size (say n), then it becomes O(n²).
````

````python
A = [1, 2, 3, 4]
B = [5, 6, 7, 8]

for i in A:
    for j in B:
        for k in range(1000000):
            print(i, j)

# Time Complexity: O(n × m), Where n = length of list A, m = length of list B
````

````python
L = [1, 2, 3, 4, 5]

for i in range(0, len(L) // 2):
    other = len(L) - i - 1
    temp = L[i]
    L[i] = L[other]
    L[other] = temp

print(L)

# Time Complexity: O(n), Where n is the number of elements in list L.
````

````python
n = 10
k = 0
for i in range(n // 2, n):
    for j in range(2, n, pow(2, j)):
        k = k + n / 2

print(k)

# Time Complexity: O(n log n) ~=O(n), Where n is the value of input n.
````

````python
a = 10
b = 3

if b <= 0:
    print(-1)
div = a // b

print(a - div - b)

# Time Complexity: O(1), Because all operations run in constant time regardless of input size.
````

````python
n = 345

sum = 0
while n > 0:
    sum = sum + n % 10
    n = n // 10

print(sum)

# Time Complexity: O(d), Where d is the number of digits in n.
````

````python
def fib(n):
    if n == 1 or n == 0:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)


# Time Complexity: O(2ⁿ), Because the recursive Fibonacci algorithm makes two calls for each non-base case, leading to exponential growth.
````

```
        = 2[2T(n-2)-1]-1 = 2^2 T(n-2) - 2 - 1

```
