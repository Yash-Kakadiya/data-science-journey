# Operators in Python

- Arithmetic Operators
- Relational Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Membership Operators

```python
# Arithmetric Operators
print(3 + 4)

print(9 - 2)

print(7 * 2)

print(14 / 2)

print(14.5 // 2)

print(7 % 3)

print(7**2)
```

```python
# Relational Operators
print(7 > 1)

print(7 < 1)

print(7 >= 7)

print(7 <= 7)

print(7 == 7)

print(7 != 7)
```

```python
# Logical Operators
print(1 and 0)

print(1 or 0)

print(not 1)
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

# bitwise not
print(~3)

# left shift and right shift
print(64 >> 3)  # 4/2^3

print(5 << 3)  # 5*2^3
```

```python
# Assignment Operators

# =
# a = 2

a = 2

# a = a % 2
a %= 2

# a++ ++a (not available in python)
a += 1

print(a)
```

```python
# Membership Operators

# in/not in

print("D" not in "Delhi")

print(1 in [2, 3, 4, 5, 6])

# Identity Operators

# is / is not
a = [1, 2, 3]
b = a
print(a is b)
```

# Loops in Python

- Need for loops
- While Loop

```python
# While loop example -> program to print the table
# Program -> Sum of all digits of a given number
# Program -> keep accepting numbers from users till he/she enters a 0 and then find the avg
```

```python
# input
number = int(input("enter the number"))
print("You entered:", number)

i = 1

# print the table
while i < 11:
    print(number, "X", i, "=", number * i)
    i += 1
```

```python
# while loop with else

x = 1

# else is executed when while condition becomes false without break
while x < 3:
    print(x)
    x += 1
else:
    print("limit crossed")
```

```python
# Guessing game

# generate a random integer between 1 and 100
import random

jackpot = random.randint(1, 100)

# input
guess = int(input("guess the number between 1 to 100:"))
print("You guessed:", guess)

counter = 1

# loop until the guess is correct
while guess != jackpot:
    if guess < jackpot:
        print("Nope! guess higher‚¨ÜÔ∏è.")
    else:
        print("Nope! guess lower‚¨áÔ∏è.")

    guess = int(input("guess again:"))
    print("You guessed:", guess)
    counter += 1
else:
    print("Correct guess‚úÖ! Number was", jackpot)
    print("Total attempts:", counter)
```

### Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
curr_pop = 10000

for i in range(10, 0, -1):
    print(i, curr_pop)
    curr_pop = curr_pop / 1.1  # x+0.1x=10000 -> x=10000/1.1
```

### Sequence sum

1/1! + 2/2! + 3/3! + ...

```python
# input
n = int(input("Enter the number:"))
print("You entered:", n)

sum = 0
fact = 1
for i in range(1, n + 1):
    fact = fact * i
    sum += i / fact
print("The sum of factorials is:", sum)
```

```python
# input
n = int(input("Enter the number of rows:"))
print("You entered rows:", n)

# loop
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end="")
    for j in range(i - 1, 0, -1):
        print(j, end="")
    print()
```

# Loop Control Statement

- Break
- Continue
- Pass

```python
# Break demo
for i in range(1, 11):
    if i == 8:
        break
    print(i)
```

```python
# Break example (Linear Search) -> Prime number in a given range
lower = int(input("Enter lower range:"))
upper = int(input("Enter upper range:"))
print("Prime numbers between", lower,"and", upper)

for num in range(lower, upper + 1):
    if num > 1:
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                break
        else:
            print(num)
```

```python
# Continue demo
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```
