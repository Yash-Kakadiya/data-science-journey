### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
# input
c = float(input("Enter temperature in Celsius: "))
print("You entered:", c, "degree C")
print("------------------------------------------")

# conversion
f = c * 9 / 5 + 32

# output
print("Temperature in Fahrenheit is:", f, "F")
````

### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
# input
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
print("You entered:", a, "and", b)
print("------------------------------------------")

print("Before swapping:")
print("First number:", a)
print("Second number:", b)
print("------------------------------------------")

# swap logic
temp = a
a = b
b = temp

# output
print("After swapping:")
print("First number:", a)
print("Second number:", b)
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
# input coordinates
x1 = int(input("Enter x coordinate first point: "))
y1 = int(input("Enter y coordinate first point: "))
x2 = int(input("Enter x coordinate second point: "))
y2 = int(input("Enter y coordinate second point: "))

print("You entered points: (", x1, ",", y1, ") and (", x2, ",", y2, ")")
print("------------------------------------------")

# distance calculation
distance = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5

# output
print("The distance between the two points is:", round(distance, 2))
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.


````python
# input
p = int(input("Enter principal amount: "))
r = float(input("Enter rate of interest: "))
t = int(input("Enter time in years: "))

print("You entered principal:", p, ", rate:", r, ", time:", t, "years")
print("------------------------------------------")

# simple interest calculation
si = (p * r * t) / 100

# output
print("The simple interest is:", round(si, 2))
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# input
heads = int(input("Enter number of heads: "))
legs = int(input("Enter number of legs: "))
print("You entered heads:", heads, "and legs:", legs)
print("------------------------------------------")

# calculation
# d+c=h => c=h-d
# 4d+2c=l
# 4d+2(h-d)=l
# 4d+2h-2d=l
# 2d+2h=l
# 2d=l-2h
# d=(l-2h)/2
dogs = (legs - 2 * heads) / 2

# c=h-d
chicken = heads - dogs

# output
print("Number of dogs:", dogs)
print("Number of cats:", chicken)
````

### Q7:- Write a program to find the sum of squares of first n natural numbers where n will be provided by the user.

````python
# input
n = int(input("Enter a number: "))
print("You entered:", n)
print("------------------------------------------")

# calculation
sum = n * (n + 1) * (2 * n + 1) / 6

# output
print("The sum of squares of first", n, "natural numbers is:", int(sum))
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# input
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
n = int(input("Enter nth term number: "))

print("You entered:", a, "and", b, "for first two terms and", n, "for nth term")
print("------------------------------------------")

print(a, b, sep=", ")

# calculation
d = b - a
res = a + (n - 1) * d

# output
print("The", n, "th term of the AP series is:", res)
````

### Q9:- Given 2 fractions, find the sum of those 2 fractions.Take the numerator and denominator values of the fractions from the user.

````python
# input
n1 = int(input("Enter numerator of first number: "))
d1 = int(input("Enter denominator of first number: "))
n2 = int(input("Enter numerator of second number: "))
d2 = int(input("Enter denominator of second number: "))

print("You entered fractions:", str(n1) + "/" + str(d1), "and", str(n2) + "/" + str(d2))
print("------------------------------------------")

# calculation
n = n1 * d2 + n2 * d1
d = d1 * d2

# output
# print("The sum of the two fractions is:", str(n)+"/"+str(d))
print(f"{n1}/{d1}+{n2}/{d2}={n}/{d}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# input
# tank
hTank = int(input("Enter height of tank in cm: "))
wTank = int(input("Enter width of tank in cm: "))
bTank = int(input("Enter breadth of tank in cm: "))

# glass
hGlass = int(input("Enter height of glass in cm: "))
rGlass = int(input("Enter radius of glass in cm: "))

print("You entered tank dimensions (h,w,b):", hTank, wTank, bTank)
print("You entered glass dimensions (h,r):", hGlass, rGlass)
print("------------------------------------------")

# calculation
volTank = hTank * wTank * bTank  # in cubic cm
volGlass = 3.14 * rGlass * rGlass * hGlass  # in cubic cm
numGlasses = volTank // volGlass

# output
print("Volume of tank in cubic cm:", round(volTank, 2))
print("Volume of glass in cubic cm:", round(volGlass, 2))
print("Number of glasses to fill the tank:", numGlasses)
````
