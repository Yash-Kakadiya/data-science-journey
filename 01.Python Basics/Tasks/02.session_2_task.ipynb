### `Problem 1`
Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Input
salary = float(input("Enter your annual salaryüí∏ (in Lakhs): "))
print("Your salary is: ‚Çπ", salary, "Lakhs")
print("-" * 50)

# Deductions
hra = 0.10 * salary
da = 0.05 * salary
pf = 0.03 * salary

# Salary after basic deductions
net_salary = salary - (hra + da + pf)

# Progressive tax calculation
tax = 0

if salary <= 5:
    tax = 0
elif salary <= 10:
    tax = (salary - 5) * 0.10
elif salary <= 20:
    tax = (5 * 0.10) + (salary - 10) * 0.20
else:
    tax = (5 * 0.10) + (10 * 0.20) + (salary - 20) * 0.30

# In-hand salary
in_hand_annual = net_salary - tax
in_hand_monthly = in_hand_annual * 100000 / 12  # Convert Lakhs ‚Üí ‚Çπ

# Output
print(f"\nTotal Tax: ‚Çπ{tax * 100000:.2f}ü•≤")
print(f"Annual In-hand Salary: ‚Çπ{in_hand_annual * 100000:.2f}üòå")
print(f"Monthly In-hand Salary: ‚Çπ{in_hand_monthly:.2f}üò¶")
````

### `Problem 5`
Display Fibonacci series up to 10 terms.

*Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34*

### `Problem 6`
Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 √ó 4 √ó 3 √ó 2 √ó 1 = 120
```

Output:

```bash

```

````python
# input
n = int(input("Enter number to find factorial:"))
fact = 1
for i in range(1, n + 1):
    fact = fact * i

print(n, "! = ", fact, sep="")
````

### `Problem 8`
Take a user input as integer N. Find out the sum from 1 to N. If any number is divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# input
n = int(input("Enter number:"))
print("You entered:", n)
print("-" * 50)

sum = 0
i = 1
while i <= n:
    if i % 5 == 0:
        i = i + 1
        continue
    if sum + i > 300:
        break
    sum += i
    i = i + 1

print("Final Sum is:", sum)
````

### `Problem 11`

### `Problem 12`
A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
print(
    """Menu:
      Type UP/DOWN/LEFT/RIGHT X where X is the number of times you want to move.
      Type ! to exit."""
)
print("-" * 50)
x = y = 0
while True:
    command = input("Enter your command: ").strip().upper()
    print("You entered:", command)
    if command == "!":
        print("Exiting the program ...")
        break
    direction, times = command.split()
    times = int(times)
    if direction == "UP":
        y += times
    elif direction == "DOWN":
        y -= times
    elif direction == "LEFT":
        x -= times
    elif direction == "RIGHT":
        x += times
    else:
        print("Invalid command. Try again.")
        continue

distance = (x**2 + y**2) ** 0.5
print("-" * 50)
print(f"Final position: ({x}, {y})")
print(f"Distance from origin: {distance:.2f}")
````

### `Problem 13`
Write a program to print whether a given number is a prime number or not.

````python
# input
n = int(input("Enter number:"))
print("You entered:", n)
print("-" * 50)

flag = True
if (n % 2) == 0:
    flag = False
for i in range(3, int(n**0.5 + 1), 2):
    if (n % i) == 0:
        flag = False
        break

if flag and n > 1:
    print(n, "is a prime numberüòÄ.")
else:
    print(n, "is not a prime numberüôÇ.")
````

### `Problem 14`
Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that equals the sum of its own digits, each raised to the power of the total number of digits. <br/>For example, 153 is an Armstrong number because 1^3+5^3+3^3==153.<br/>
0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
lower = int(input("Enter lower range:"))
upper = int(input("Enter upper range:"))
print("You entered range:", lower, "to", upper)
print("-" * 50)
print("Armstrong numbers in the given range are:")
for num in range(lower, upper + 1):
    len_num = 0
    temp = num
    while temp > 0:
        len_num += 1
        temp = temp // 10
    sum = 0
    temp = num
    while temp > 0:
        digit = temp % 10
        sum += digit**len_num
        temp = temp // 10
    if sum == num:
        print(num, end=" ")
````

### `Problem 15`
Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
# input
hour = int(input("Enter hours:"))
minute = int(input("Enter minutes:"))
print("You entered time:", hour, "hours and", minute, "minutes")
print("-" * 50)

hour_angle = (hour % 12) * 30 + (minute / 60) * 30
minute_angle = minute * 6
angle = abs(hour_angle - minute_angle)
if angle > 180:
    angle = 360 - angle

print(f"The angle between hour and minute hands is: {angle} degrees")
````
